```python
import cv2
from google.colab.patches import cv2_imshow

image_path = '/image 4.jfif'

# Read the image using cv2.imread()
img = cv2.imread(image_path)

# Check if the image was loaded succesfully
if img is None:
  print(f"Error:image not found at {image_path} or unable to load.")
else:
  # Dislay the image using cv2_imshow
  cv2_imshow(img)
```
